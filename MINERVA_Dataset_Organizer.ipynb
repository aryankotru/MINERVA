```python
import scipy.io
import pandas as pd

# Load EEG data from the .mat file
file_path = 'S001.mat'
eeg_data = scipy.io.loadmat(file_path)['data']

# Extract relevant information
channel_names = ['POz', 'PO3', 'PO4', 'PO5', 'PO6', 'Oz', 'O1', 'O2']
electrode_labels = ['dry', 'wet']
block_indices = 10
target_indices = 12

# Reshape the 5D matrix into a 2D matrix (Time points x Channels)
num_channels, num_time_points, num_electrodes, num_blocks, num_targets = eeg_data.shape
eeg_2d = eeg_data.reshape(num_channels, -1).T

# Create a DataFrame with appropriate column labels
columns = []
for electrode_label in electrode_labels:
    for channel_name in channel_names:
        for block_index in range(num_blocks):
            for target_index in range(num_targets):
                columns.append(f'{channel_name}_{electrode_label}_B{block_index + 1}_T{target_index + 1}')


df = pd.DataFrame(data=eeg_2d, columns=columns).loc[~df.index.duplicated()]


# Print the DataFrame
print(df)
```
